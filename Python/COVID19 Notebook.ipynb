```python
# import the modules
import pandas as pd 
import numpy as np 
import seaborn as sns
import matplotlib
%matplotlib inline
# adjust the configuration of the plots we will create
matplotlib.rcParams['figure.figsize'] = (12,8) 
import matplotlib.pyplot as plt 

import warnings
warnings.filterwarnings('ignore')
```

```python
# import the dataset
covid_df = pd.read_csv("covid19_Confirmed_dataset.csv")
# preview
covid_df.head(10)
```

```python
# check the shape of the dataframe
covid_df.shape
```

```python
# remove unused columns
covid_df.drop(['Lat', 'Long'], axis=1, inplace = True)
covid_df.head(10)
```

### Task 2.3: Aggregating the rows by the country

```python
# Aggregating the rows by the country
covid_df1 = covid_df.groupby("Country/Region").sum()
covid_df1.head(10)
```

```python
# visualizing data filtered by country
covid_df1.loc['China'].plot()
covid_df1.loc['Italy'].plot()
covid_df1.loc['Spain'].plot()
covid_df1.loc['United Kingdom'].plot()
plt.legend()
```

```python
# decribing the spread of the virus in a country
covid_df1.loc['United Kingdom'].plot()
```

```python
# calculating the first derivative of the curve
covid_df1.loc['United Kingdom'].diff().plot()
```

```python
# find max infection rate for United Kingdom
covid_df1.loc['United Kingdom'].diff().max()
```

```python
# find the max infection rate for all countries in the dataframe
countries = list(covid_df1.index)
max_infection_rates = []
for country in countries:
    max_infection_rates.append(covid_df1.loc[country].diff().max())
    
# create new 'max infection rate' column to the dateframe
covid_df1['max infection rate'] = max_infection_rates
```

```python
covid_df1.head()
```

```python
# view the max infection rate of every country/region
covid_df2 = pd.DataFrame(covid_df1['max infection rate'])
covid_df2.head(10)
```

```python
# import the dataset
world_happiness_df = pd.read_csv("worldwide_happiness_report.csv")
world_happiness_df.head(10)
```

```python
# remove unused columns
columns_to_dropped = ['Overall rank','Score','Generosity','Perceptions of corruption']
world_happiness_df = world_happiness_df.drop(columns_to_dropped,axis=1)
world_happiness_df.head(10)
```

```python
# change the indices of the dateframe
world_happiness_df = world_happiness_df.set_index(['Country or region'])
world_happiness_df.head()
```
